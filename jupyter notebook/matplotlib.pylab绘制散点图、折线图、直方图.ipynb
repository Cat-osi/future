```python
import matplotlib.pylab as pyl
import numpy as np
x=np.random.random_integers(2,20,1000)
y=np.random.normal(15,6,1000)#创建正态分布数据
pyl.subplot(2,1,1)
pyl.plot(x,y,'og')   #散点图
pyl.subplot(2,1,2)
pyl.hist(y,np.arange(1,100,1.5))#绘制直方图 统计一个序列用的
pyl.show()
```
