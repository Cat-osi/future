# 1.离差标准化 （最小-最大标准化）
##### 消除量纲（单位）影响以及变异大小因素的影响
#####  x1=（x-min）/（max-min）

```python
import pandas as pd
import numpy as np
data=pd.read_csv('C:\\Users\\ASUS\\Desktop\\python\\chapter4\\classification.csv')
#离差标准化
data1=(data-data.min())/(data.max()-data.min())
data1
```

# 2.标准差标准化 （零均值标准化）
##### 消除单位影响以及变量自身变异影响
#####  x1=（x-平均数）/标准差

```python
#data.mean()获取平均数  data.std()获取标准差
data2=(data-data.mean())/data.std()
data2
```

# 3.小数定标规范化
##### 消除单位影响
#####  x1=x/10**（k）
#####  k=log10（x的绝对值的最大值）

```python
k=np.log10(data.abs().max())
data3=data/10**k
data3
```
